```python
import sqlite3
import matplotlib.pyplot as plt
import numpy


def read_performance_data(filename):
    connection = sqlite3.connect(filename)
    cursor = connection.cursor()
    
    cursor.execute("SELECT message FROM log_table WHERE category = 'DensityGridOverlay'")
    
    return list(cursor)

data = read_performance_data("perf_db")

devices = [int(p[0].split(" ")[0]) for p in data]
duration = [int(p[0].split(" ")[1]) for p in data]

_, ax = plt.subplots(1, figsize = (8, 4))

ax.plot(devices, duration, 'o', color='blue', markersize=3)
ax.set_ylabel('Duration in ms')
ax.set_xlabel('Number of cells')

pass
```
